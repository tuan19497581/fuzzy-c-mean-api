```python
import pandas as pd
import numpy as np
```

```python
filepath = '../data/9_CNPM.xls'
```

```python
df = pd.read_excel(filepath)
```

```python
df.head(10)
```

```python
mon_tu_van = [
    'nhập môn lập trình',
    'toán rời rạc',
    'khai thác dữ liệu',
    'cơ sở dữ liệu',
    'hệ quản trị cơ sở dữ liệu',
    'cấu trúc dữ liệu và giải thuật',
    'lập trình hướng đối tượng',
    'kiến trúc máy tính',
    'mạng máy tính',
    'hệ điều hành'
]
```

```python
data = {}
```

# KhoaDaoTao

```python
ten_kdt = df.iloc[3, 12].split(':')[-1].strip()
ten_kdt
```

```python
data['KhoaDaoTao'] = dict(ten_kdt=ten_kdt)
```

# ChuyenNganh

```python
ten_cn = df.iloc[5, 1].split(':')[-1].strip()
ten_cn
```

```python
data['ChuyenNganh'] = dict(ten_cn=ten_cn)
```

# ChuongTrinhDaoTao

```python
all_subjects = df.iloc[7, 5:-6].values
all_subjects.shape
```

```python
tin_chi = np.array(list(map(lambda x: int(x.strip()), df.iloc[8, 5:-6].values)))
tin_chi.shape
```

```python
list_ctdt = []

for subj, tc in zip(all_subjects, tin_chi):
    subj = subj.lower()
    idx = 0
    if subj in mon_tu_van:
        idx = mon_tu_van.index(subj) + 1
        
    ctdt = dict(
        stt_monhoc=idx,
        ten_mh=subj,
        tin_chi=tc
    )
    list_ctdt.append(ctdt)
```

```python
len(list_ctdt)
```

```python
data['ChuongTrinhDaoTao'] = list_ctdt
```

# SinhVien

```python
ho_dem, ten = df.iloc[9:-2, 2:4].values.T
ho_dem.shape, ten.shape
```

```python
list_sv = []
for hd, t in zip(ho_dem, ten):
    sv = dict(ho_dem_sv=hd, ten_sv=t)
    list_sv.append(sv)
```

```python
data['SinhVien'] = list_sv
```
